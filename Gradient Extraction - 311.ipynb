```python
import numpy as np
import cv2
import pylab as pyl

# Load the image and display
image = cv2.imread(r"C:\Users\HP\Desktop\DIP Assignment 1\Assignment 2\Image_a2.tif",0)
cv2.imshow("Original",image)
pyl.imshow(image)
```

```python
# Compute the Laplacian of the image
lap = cv2.Laplacian(image, cv2.CV_64F)
lap = np.uint8(np.absolute(lap))
cv2.imshow("Laplacian", lap)
pyl.imshow(lap)
```

The Laplacian is used to measure the 2nd spatial derivative of an image. It highlights the regions of rapid intensity change. It sharpens the edges of the image. It is useful for finding the fine details of an image. Any feature with a sharp discontinuity will be enhanced by a Laplacian operator.

```python
# Gradients along the X axis
sobelX = cv2.Sobel(image, cv2.CV_64F, 1, 0)
sobelX = np.uint8(np.absolute(sobelX))
cv2.imshow("Sobel X", sobelX)
pyl.imshow(sobelX)
```

A gradient image in the x direction measuring horizontal change in intensity.

```python
# Gradients along the Y axis
sobelY = cv2.Sobel(image, cv2.CV_64F, 0, 1)
sobelY = np.uint8(np.absolute(sobelY))
cv2.imshow("Sobel Y", sobelY)
pyl.imshow(sobelY)
```

A gradient image in the y direction measuring vertical change in intensity.

Each pixel of the gradient image measures the change in intensity of that same point in the original image, in a given direction.The pixels with the largest gradient values in the direction of the gradient become edge pixels.

```python
# Gradients along the X and Y axis combined
sobelCombined = cv2.bitwise_or(sobelX, sobelY)

cv2.imshow("Sobel Combined", sobelCombined)
pyl.imshow(sobelCombined)
```

To get the full range of direction, gradient images in the x and y directions are computed and combined. The results of the X and Y gradient images are combined by applying a logical bitwise OR to produce a resulting combined Sobel image.
